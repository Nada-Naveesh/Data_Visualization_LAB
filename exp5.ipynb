```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("25-42700.csv")
plt.figure(figsize=(7,5))
sns.scatterplot(data=df,x="sepal_length",y="sepal_width",hue="species")
plt.title("Simple Scatter Plot: Sepal Length vs Sepal Width")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(data=df, x="petal_length", y="petal_width", hue="species", size="sepal_length", sizes=(20,200),alpha=0.7)
plt.title("MultivariateScatterPlot")
plt.show()
```

```python
sns.pairplot(df,hue="species",diag_kind="hist")
plt.suptitle("Scatter Plot Matrix of Iris Dataset", y=1.02)
plt.show()
```
