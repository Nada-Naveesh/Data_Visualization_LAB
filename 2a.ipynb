```python
#importing the required libraries

import pandas as pd
import matplotlib.pyplot as plt
```

```python
#reading the dataset

df=pd.read_csv("exp2a.csv")
print(df)
```

```python

if df.shape[1] >= 2:
  plt.figure(figsize=(6,5))
  for i in range(len(df)):
    plt.plot([0,1],[df.iloc[i,0],df.iloc[i,1]],marker="o")
  plt.xticks([0,1],[df.columns[0],df.columns[1]])
  plt.title("Slope Chart - First 2 columns")
  plt.show()
```
