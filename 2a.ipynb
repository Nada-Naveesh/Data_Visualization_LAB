```python
#importing the required libraries

import pandas as pd
import matplotlib.pyplot as plt
```

```python
#reading the dataset

df=pd.read_csv("exp2a.csv")
print(df)
```

```python

if df.shape[1] >= 2:
  plt.figure(figsize=(6,5))
  for i in range(len(df)):
    plt.plot([0,1],[df.iloc[i,0],df.iloc[i,1]],marker="o")
  plt.xticks([0,1],[df.columns[0],df.columns[1]])
  plt.title("Slope Chart - First 2 columns")
  plt.show()
```

```python
#comparission
df.plot(kind="bar",figsize=(8,5),width=0.8)
plt.title("comaprison chart")
plt.ylabel("death rate")
plt.xlabel("index (age group)")
plt.legend(title="category")
plt.show()
```

```python
import matplotlib.pyplot as plt
values = df.select_dtypes(include="number").sum().sort_values(ascending=False)
cum_percent = values.cumsum() / values.sum() * 100
fig, ax = plt.subplots()
ax.bar(values.index, values)
ax2 = ax.twinx()
ax2.plot(values.index, cum_percent, color="red", marker="D")
ax.set_ylabel("Frequency")
ax2.set_ylabel("Cumulative Percentage")
plt.title("Pareto Chart")
plt.xticks(rotation=45)
plt.show()
```

```python

import pandas as pd
import plotly.express as px



fig = px.bar_polar(
    r=df.iloc[6].values,
    theta=df.columns,
    template="plotly_dark"
)

fig.update_layout(title="Radial Bar Chart")
fig.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

categories = list(df.columns)
values = df.iloc[0].values.flatten().tolist()
values += values[:1]  # repeat first value to close the circle

angles = np.linspace(0, 2*np.pi, len(categories), endpoint=False).tolist()
angles += angles[:1]  # repeat first angle

plt.figure(figsize=(10,10))
ax = plt.subplot(111, polar=True)

# plot line
ax.plot(angles, values, marker="o")
# fill area
ax.fill(angles, values, alpha=0.25)

# set category labels
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories)

plt.title("Radar Chart")
plt.show()
```
