```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# 1. Load Dataset
df = pd.read_csv("exp3b.csv")
# 2. Create Date Column (Assuming year 1973)
df['Date'] = pd.to_datetime({
    'year': 1973,
    'month': df['Month'],
    'day': df['Day']
})
# 3. Visualization Settings
sns.set(style="whitegrid")
# 4. Ozone Over Time
plt.figure(figsize=(12,6))
plt.plot(df['Date'], df['Ozone'], color='blue')
plt.title("Ozone Concentration Over Time (1973)")
plt.xlabel("Date")
plt.ylabel("Ozone (ppb)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
# 5. Ozone Distribution
plt.figure(figsize=(8,5))
sns.histplot(df['Ozone'].dropna(), bins=20, kde=True)
plt.title("Distribution of Ozone Levels")
plt.xlabel("Ozone (ppb)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
# 6. Monthly Average Ozone
monthly_avg = df.groupby('Month')['Ozone'].mean()
plt.figure(figsize=(8,5))
monthly_avg.plot(kind='bar', color='orange')
plt.title("Average Monthly Ozone Levels")
plt.xlabel("Month")
plt.ylabel("Average Ozone (ppb)")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
