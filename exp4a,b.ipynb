```python
#exp4a

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# 1. Load Iris Dataset
df = pd.read_csv("iris.csv")
# 2. Visualization Settings
sns.set(style="whitegrid")
palette = sns.color_palette("Set2")  # Colors for species
# 3. Melt the DataFrame for Combined Box Plot
numeric_cols = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
df_melt = df.melt(id_vars='species', value_vars=numeric_cols,
                  var_name='Feature', value_name='Value')
# 4. Combined Box Plot
plt.figure(figsize=(12,6))
sns.boxplot(x='Feature', y='Value', hue='species', data=df_melt, palette=palette)
plt.title("Iris Features Distribution by Species")
plt.xlabel("Feature")
plt.ylabel("Value")
plt.legend(title='Species')
plt.tight_layout()
```

```python
#exp4b

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# 1. Load Air Quality Dataset
df = pd.read_csv("exp3b.csv")
# 2. Visualization Settings
sns.set(style="whitegrid")
palette = sns.color_palette("Set2")  # Colors for months
# 3. Melt the DataFrame for Combined Box Plot
parameters = ['Ozone', 'Solar.R', 'Wind', 'Temp']
df_melt = df.melt(id_vars='Month', value_vars=parameters,
                  var_name='Parameter', value_name='Value')
# 4. Combined Box Plot
plt.figure(figsize=(12,6))
sns.boxplot(x='Parameter', y='Value', hue='Month', data=df_melt, palette=palette)
plt.title("Air Quality Parameters Distribution by Month")
plt.xlabel("Parameter")
plt.ylabel("Value")
plt.legend(title='Month', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
