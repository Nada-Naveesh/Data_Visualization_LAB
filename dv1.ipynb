```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("dv1.csv")

print("VAdeaths")
print(df)

df.plot(kind='line',marker='o',figsize=(8,5))
plt.title("--VAdeaths--")
plt.xlabel("Age Group")
plt.ylabel("Death Rate")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df.plot(kind='bar',figsize=(8,5))
plt.title("--VAdeaths--")
plt.xlabel("Age Group")
plt.ylabel("Death Rate")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(df['Rural Male'], df['Rural Female'], color='blue', label='Rural Male vs Rural Female')
plt.title('Scatter Plot of Rural Male vs. Rural Female Death Rates')
plt.xlabel('Rural Male Death Rate')
plt.ylabel('Rural Female Death Rate')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(df['Urban Male'], df['Urban Female'], color='blue', label='Rural Male vs Rural Female')
plt.title('Scatter Plot of Urban Male vs. Uraban Female Death Rates')
plt.xlabel('Urban Male Death Rate')
plt.ylabel('Urban Female Death Rate')
plt.grid(True)
plt.legend()
plt.show()
```
